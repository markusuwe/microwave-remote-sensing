# Electromagnetic Radiation as Waves

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To get an intuition, let's use the wavenumber describing Sentinel-1 C-band data
# The Sentinel-1 instrument uses a frequency of 5.405 GHz
freq = 5.405
# Convert to wavelength in cm
wavelength = 29.9792458 / freq
print(f"The wavelength for Sentinel-1 is {wavelength:.3f} cm")
```

```python
A = 0.8      # Amplitude
k = 2 * np.pi / wavelength # Formula to derive wavenumber from wavelength
z = np.arange(0.0, 10.0, 0.01)
psi = A * np.sin(k * z)
```

```python
# size chosen so that it comes close to cm units, but depends on display
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(z, psi)
```

```python
# ICEYE uses instruments with 9.65 GHz (X-band)
freq = 9.65
wavelength = 29.9792458 / freq
print(f"The wavelength for ICEYE is {wavelength:.3f} cm")
```

```python
A = 2
k = 2 * np.pi / wavelength
z = np.arange(0.0, 10.0, 0.01)
psi = A * np.sin(k * z)
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(z, psi)
```
